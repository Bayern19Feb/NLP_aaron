```python
text = open("story.txt",encoding="utf-8").read()

text_lower = text.lower()
```

```python
text_lower
```

```python
# Displayes all the punctuations in the text.
string.punctuation
```

```python
# Removing any kind of punctuations from the text.
cleaned_text = text_lower.translate(str.maketrans("", "",string.punctuation))
```

```python
cleaned_text
```

```python
# Using word_tokenize to tokenize sentence into words
```

```python
nltk.download('punkt')
```

```python
tokenized_words = word_tokenize(cleaned_text, "english")
```

```python
tokenized_words
```

```python
len(tokenized_words)
```

```python
# Removing stop words
final_words = []

for word in tokenized_words:
    
    if word not in stopwords.words('english'):
        
        final_words.append(word)
```

```python
final_words
```

```python
len(final_words)
```

```python
lemma_words = []

for word in final_words:
    
    word = WordNetLemmatizer().lemmatize(word)
    
    lemma_words.append(word)
```

```python
lemma_words
```

```python
len(lemma_words)
```

```python
emotion_list = []

with open("emotions.txt","r") as file:
    
    for line in file:
        
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        
        word, emotion = clear_line.split(':')
        
        if word in lemma_words:
            
            emotion_list.append(emotion)
            
print("People emotions from the text \n", emotion_list, '\n \n')
```

```python
w = Counter(emotion_list)
print("Count of each emotion \n", w)
```

#### VADER

VADER ( Valence Aware Dictionary for Sentiment Reasoning) is a model used for text sentiment analysis that is sensitive to both polarity (positive/negative) and intensity (strength) of emotion. It is available in the NLTK package and can be applied directly to unlabeled text data. VADER sentimental analysis relies on a dictionary that maps lexical features to emotion intensities known as sentiment scores. The sentiment score of a text can be obtained by summing up the intensity of each word in the text. For example- Words like ‘love’, ‘enjoy’, ‘happy’, ‘like’ all convey a positive sentiment. Also VADER is intelligent enough to understand the basic context of these words, such as “did not love” as a negative statement. It also understands the emphasis of capitalization and punctuation, such as “ENJOY”

## Polarity classification

We won’t try to determine if a sentence is objective or subjective, fact or opinion. Rather, we care only if the text expresses a positive, negative or neutral opinion.

```python
# Creating a function
def sentiment_analyse(sentiment_text):
    
    
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    
    
    if score['neg'] > score['pos']:
        print("\n     ******Negative Sentiment*******")
        
    
    elif score['neg'] < score['pos']:
        print("\n     ******Positive Sentiment*******")
        
    
    else:
        print("Neutral Sentiment")
```

```python
sentiment_analyse("a very bad movie")
```

```python
sentiment_analyse("best movie i have ever seen")
```
